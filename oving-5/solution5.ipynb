```python
# Task 1, 2, 3
import pandas as pd
from sklearn import datasets

df = pd.read_csv('./agaricus-lepiota.data')
df
```

```python
# Task 4/5
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from sklearn.preprocessing import LabelEncoder

# First we must convert the data into ordinal data
label_encoder = LabelEncoder()

df.astype('category')
for column in df.columns:
    df[column] = label_encoder.fit_transform(df[column])

# Remove veil-type from the dataset, as it only has one unique value
if "veil-type" in df.columns: df = df.drop(["veil-type"], axis=1) 

df.head()
```

```python
# Show distributions
plt.figure()
df.hist(color='k', alpha=0.5, bins=10, figsize=(12, 12))
```

```python
scatter_matrix(df, figsize=(12, 12), diagonal='kde')
```

```python
# Task 6
import numpy as np
df = pd.read_csv('./agaricus-lepiota.data')
df = pd.get_dummies(df)
dummies = pd.get_dummies(df).rename(columns=lambda x: 'Category_' + str(x))
df = pd.concat([df, dummies], axis=1)
print(df)


indices = np.argsort(df.columns)

plt.spy(df, markersize=0.1)
fig = plt.gcf()
fig.set_size_inches(200, 200)
plt.plot()
plt.show()
```
