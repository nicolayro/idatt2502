```python
# Task 1, 2, 3
import pandas as pd
from sklearn import datasets

df = pd.read_csv('./agaricus-lepiota.data')
df
```

```python
# Task 4/5
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from sklearn.preprocessing import LabelEncoder

# First we must convert the data into ordinal data
label_encoder = LabelEncoder()

df.astype('category')
for column in df.columns:
    df[column] = label_encoder.fit_transform(df[column])

# Remove veil-type from the dataset, as it only has one unique value
if "veil-type" in df.columns: df = df.drop(["veil-type"], axis=1) 

df.head()
```

```python
# Show distributions
plt.figure()
df.hist(color='k', alpha=0.5, bins=10, figsize=(12, 12))
```

```python
scatter_matrix(df, figsize=(12, 12), diagonal='kde')
```

```python
# Task 6
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
vectorizer = TfidfVectorizer()
#vectorizer = CountVectorizer()
vectors = vectorizer.fit_transform(df)


#vectorizer = CountVectorizer()

# Sort: indices now can be used to sort by newsgroup below
indices = np.argsort(df.columns)

# Count word frequencies throughout all documents
freqs = vectors[indices].sum(axis=0)

# Sort word frequencies
i = np.argsort(freqs)
vocab = pd.Series(vectorizer.vocabulary_)
print(vocab[vocab.values==i[...,-1].item()])

# Plot document-term matrix, blue indicates non-zero value
plt.spy(vectors[indices], markersize=100)
fig = plt.gcf()
fig.set_size_inches(60,60)
plt.plot()
plt.show()
```
